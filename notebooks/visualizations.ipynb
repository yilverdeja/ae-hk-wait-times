```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the mapping from wait descriptions to numerical values
wait_mapping = {
    'Around 1 hour': 1,
    'Over 1 hour': 2,
    'Over 2 hours': 3,
    'Over 3 hours': 4,
    'Over 4 hours': 5,
    'Over 5 hours': 6,
    'Over 6 hours': 7,
    'Over 7 hours': 8,
    'Over 8 hours': 9
}
```

```python
# Load the data
df = pd.read_csv('../data/raw_data.csv', parse_dates=['Timestamp'], index_col=['Timestamp'])
```

```python
# Convert wait times using the mapping
for column in df.columns:  # Skip the first column (Timestamp)
    df[column] = df[column].map(wait_mapping)
```

```python
df.resample(rule='ME').mean().plot(figsize=(12,6)).legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
```

```python
df.resample(rule='W').mean().plot(figsize=(12,6)).legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
```

```python
for column in df.columns:
    plt.figure(figsize=(12, 6))  # Create a new figure with a defined size

    # Daily mean
    daily_mean = df[column].resample('D').mean()
    daily_mean.plot(label='Daily Mean')

    # Weekly mean
    weekly_mean = df[column].resample('W').mean()
    weekly_mean.plot(label='Weekly Mean')

    # Monthly mean
    monthly_mean = df[column].resample('ME').mean()
    monthly_mean.plot(label='Monthly Mean')

    # Adding plot titles and labels
    plt.title(f'Means for {column}')
    plt.xlabel('Date')
    plt.ylabel('Values')
    plt.grid(True)

    # Adding legend
    plt.legend()

    # Show plot
    plt.show()
```
